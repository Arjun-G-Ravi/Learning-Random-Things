Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Loading Image

```python
path = 'Peacock.jpg' # Path to image
image = cv2.imread(path)
```

Printing using matplotlib

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

# Drawing objects

- cv2.line(image, startPoint, endPoint, rgb, thinkness)
- cv2.rectangle(image, topLeft, bottomRight, rgb, thinkness)
- cv2.circle(image, center, radius, rgb, thinkness)
- cv2.ellipse(image, center, axes, angle, startAngle, endAngle, rgb, thinkness)

NOTE: Negative thickness fills the image with the specified color

Line

```python
img_line = cv2.line(image, (100,857), (920,50), (255, 255,0), 10)
plt.imshow(cv2.cvtColor(img_line, cv2.COLOR_BGR2RGB))
```

```python
img_circ = cv2.circle(image, (600,657), 300, (255,9,0), -10)
plt.imshow(cv2.cvtColor(img_circ, cv2.COLOR_BGR2RGB))
```

```python
img_rect = cv2.rectangle(image, (100,857), (920,50), (255,0,255), 10)
plt.imshow(cv2.cvtColor(img_rect, cv2.COLOR_BGR2RGB))
```

```python
# cv2.ellipse(image, center, axes, angle, startAngle, endAngle, rgb, thinkness)

img_ellipse = cv2.ellipse(image,(256,256), (100,50), -45, 0, 360, (255,0,0), 10)   
plt.imshow(cv2.cvtColor(img_ellipse, cv2.COLOR_BGR2RGB))
```
